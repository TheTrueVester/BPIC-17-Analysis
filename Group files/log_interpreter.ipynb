```python
import pm4py
import pandas as pd

file_path = os.path.join("data", "BPI_Challenge_2017.xes.gz")
log = pm4py.read_xes(file_path)
dataframe = pm4py.convert_to_dataframe(log)
print(dataframe.head())
print(dataframe.info())
```

```python
nr_cases = dataframe['case:concept:name'].nunique()
nr_events = len(dataframe)
nr_activities = dataframe['concept:name'].nunique()
nr_resources = dataframe['org:resource'].nunique()
log_variants = pm4py.get_variants_as_tuples(log)
nr_variants = len(log_variants)
dominating_activities = dataframe['concept:name'].value_counts()
print("-- Basic Event Log Properties: suggested in the lecture/Task 3 ---")
print(f"Total Cases: {nr_cases:,}")
print(f"Unique Resources: {nr_resources}")
print(f"Total Process Variants: {nr_variants:,}")
print(f"Total Events Recorded: {nr_events:,}")
print(f"Unique Activities: {nr_activities}")
print(f"Activity Frequency Distribution: {dominating_activities}")
```

```python
# Average Cycle Times of Cases
import numpy
from datetime import timedelta 
import pandas as pd

case_durations = dataframe.groupby('case:concept:name')['time:timestamp'].agg(['min','max'])

case_durations['duration'] = case_durations['max'] - case_durations['min']

duration_seconds = case_durations['duration'].dt.total_seconds()
avg_duration_seconds = duration_seconds.mean()
std_dev_duration_seconds = duration_seconds.std()

avg_duration_formatted = timedelta(seconds=avg_duration_seconds)
std_duration_formatted = timedelta(seconds=std_dev_duration_seconds)

case_lengths = dataframe.groupby('case:concept:name').size()
mean_case_length = case_lengths.mean ()
std_dev_case_length = case_lengths.std()

print("\n Cycle Time & Case Length ")
print(f"Cycle Time: {avg_duration_formatted}")
print(f"Standard Deviation of Duration: {std_duration_formatted}")
print(f"Mean Case Length (Events per Case): {mean_case_length: .2f}")
print(f"Standard Deviation of Case Length: {std_dev_case_length: .2f}")
```

```python
#Heuristics Miner
from pm4py.algo.discovery.heuristics import algorithm as heuristics_miner
from pm4py.visualization.petri_net import visualizer as pn_visualizer

parameters = {"dependency_threshold": 0.9}

result_tuple = heuristics_miner.apply(log, parameters=parameters)


petri_net, initial_marking, final_marking = result_tuple

gviz = pn_visualizer.apply(petri_net, initial_marking, final_marking)
pn_visualizer.save(gviz, "heuristics_miner_petri.png")

print("Heuristics Miner Model Generated.")
```

```python
#Alpha Miner
from pm4py.algo.discovery.alpha import algorithm as alpha_miner
from pm4py.visualization.petri_net import visualizer as pn_visualizer


petri_net_alpha, initial_marking_alpha, final_marking_alpha = alpha_miner.apply(log)


gviz_alpha = pn_visualizer.apply(petri_net_alpha, initial_marking_alpha, final_marking_alpha)
pn_visualizer.save(gviz_alpha, "alpha_miner_petri.png")

print("Alpha Miner Petrinet Generated.")

```

```python
from pm4py.algo.evaluation.replay_fitness import algorithm as replay_fitness
from pm4py.algo.evaluation.precision.variants import etconformance_token as precision_token
from pm4py.algo.evaluation.generalization import algorithm as generalization_algo
from pm4py.algo.evaluation.generalization.variants import token_based as generalization_token
from pm4py.algo.evaluation.simplicity import algorithm as simplicity_evaluator



fitness_h = replay_fitness.apply(log, petri_net, initial_marking, final_marking, variant=replay_fitness.Variants.TOKEN_BASED)
simplicity_nodes_h = len(petri_net.places) + len(petri_net.transitions)
simplicity_arcs_h = len(petri_net.arcs)

precision_h = precision_token.apply(log, petri_net, initial_marking, final_marking)

generalization_h = generalization_algo.apply(
    log, petri_net, initial_marking, final_marking, variant=generalization_algo.Variants.GENERALIZATION_TOKEN
)
simplicity_h = simplicity_evaluator.apply(petri_net)

print("Heuristics Miner Metrics:")
print(f"Fitness: {fitness_h}")
print(f"Simplicity - Nodes: {simplicity_nodes_h}, Arcs: {simplicity_arcs_h}")
print(f"Precision: {precision_h:.4f}")
print(f"Generalization: {generalization_h:.4f}")
```

```python
from pm4py.algo.filtering.log.attributes import attributes_filter

end_activities = [
    "A_Cancelled",
    "A_Denied",
    "A_Pending",
    "O_Cancelled",
    "O_Refused"
]


filtered_log = attributes_filter.apply_events(
    log,
    values=end_activities,
    parameters={
        attributes_filter.Parameters.ATTRIBUTE_KEY: "concept:name",
        attributes_filter.Parameters.POSITIVE: True
    }
)

print(f"Original log cases: {len(log)}")
print(f"Filtered log cases: {len(filtered_log)}")
```

```python
from pm4py.algo.discovery.heuristics import algorithm as heuristics_miner
from pm4py.visualization.petri_net import visualizer as pn_visualizer

parameters = {"dependency_threshold": 0.9}
result_tuple = heuristics_miner.apply(filtered_log, parameters=parameters)
net, im, fm = result_tuple

gviz = pn_visualizer.apply(net, im, fm)
pn_visualizer.save(gviz, "heuristics_miner_filtered.png")
```
